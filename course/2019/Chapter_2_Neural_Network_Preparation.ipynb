```python
import numpy as np
from sklearn import linear_model
X_train = np.array([[1],[2],[3],[4]])
Y_train = np.array([1,2,3,4])
model = linear_model.LinearRegression()
model.fit(X_train, Y_train)
coef1 = model.coef_[0]
intercept1 = model.intercept_
print(model.predict([[7]]))
```

```python
from sklearn import linear_model
X_train = np.array([[1],[2],[3],[4]])
Y_train = np.array([3,4,5,6])
model = linear_model.LinearRegression()
model.fit(X_train, Y_train)
coef2 = model.coef_[0]
intercept2 = model.intercept_
print(model.predict([[7]]))
```

```python
X_train = np.array([[0],[0.1],[2],[2.1],[4],[4.1]])
Y_train = np.array([0,2,1,3,2,4])
model = linear_model.LinearRegression()
model.fit(X_train, Y_train)
coef3 = model.coef_[0]
intercept3 = model.intercept_
print(model.predict([[4]]))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 4))
x = np.linspace(-5.0, 5.0)

y1 = coef1*x + intercept1
X1_train = [1,2,3,4]
Y1_train = [1,2,3,4]

y2 = coef2*x + intercept2
X2_train = [1,2,3,4]
Y2_train = [3,4,5,6]

y3 = coef3*x + intercept3
X3_train = [0, 0.1, 2, 2.1, 4, 4.1]
Y3_train = [0, 2, 1, 3, 2, 4]

plt.subplot(1, 3, 1)
plt.plot(x, y1, c='black', lw=3)
plt.scatter(X1_train, Y1_train, marker='X', s=200);
plt.axhline(lw=1, c='black')
plt.axvline(lw=1, c='black')
plt.grid(alpha=0.4, ls='-.')

plt.subplot(1, 3, 2)
plt.plot(x, y2, c='black', lw=3)
plt.scatter(X2_train, Y2_train, marker='X', s=200);
plt.axhline(lw=1, c='black')
plt.axvline(lw=1, c='black')
plt.grid(alpha=0.4, ls='-.')


plt.subplot(1, 3, 3)
plt.plot(x, y3, c='black', lw=3)
plt.scatter(X3_train, Y3_train, marker='X', s=200);
plt.axhline(lw=1, c='black')
plt.axvline(lw=1, c='black')
plt.grid(alpha=0.4, ls='-.')

plt.show()
```

```python
from sklearn import linear_model
from random import random, seed

n = 50
seed(1)                                     
X_train = [[random(), random()] for i in range(n)]              
Y_train = [random() for i in range(n)]

model = linear_model.LinearRegression()
model.fit(X_train, Y_train)
coef_x1 = model.coef_[0]
coef_x2 = model.coef_[1]
intercept = model.intercept_
print(model.predict([[7,8]]))
```

```python
from mpl_toolkits.mplot3d import *
import matplotlib.pyplot as plt
import numpy as np
from random import random, seed
from matplotlib import cm


fig = plt.figure()
ax = fig.gca(projection='3d')               # to work in 3d
# plt.hold(True)

x1_surf=np.arange(0, 1, 0.01)                # generate a mesh
x2_surf=np.arange(0, 1, 0.01)
x1_surf, x2_surf = np.meshgrid(x1_surf, x2_surf)
y_surf = coef_x1* x1_surf + coef_x2 * x2_surf + intercept 
ax.plot_surface(x1_surf, x2_surf, y_surf, cmap=cm.hot);    # plot a 3d surface plot

n = 100
seed(0)                                     # seed let us to have a reproducible set of random numbers
x1 = [x[0] for x in X_train]
x2 = [x[1] for x in X_train]

ax.scatter(x1, x2, Y_train);                        # plot a 3d scatter plot

ax.set_xlabel('x1 feature1')
ax.set_ylabel('x2 feature2')
ax.set_zlabel('y label/target')

plt.show()
```

## Non-Linear Functions
$$
sigmoid(x)={\frac {1}{1+e^{-x}}}
$$

$$
tanh(x)={\frac {e^{x} - e^{-x}}{e^{x}+e^{-x}}}
$$

$$
ReLU(x) = \left\{
        \begin{array}{lll}
            0 & for & x \leq 0  \\
            x & for & x > 0
        \end{array}
    \right.
$$

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-5.0, 5.0)

y1 = 1 / (1 + np.exp(-x))
y2 = (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x)) 
y3 = np.where(x>0,x,0.0)

plt.figure(figsize=(20, 4))

plt.subplot(1, 3, 1)
plt.plot(x, y1, c='black', lw=3)
plt.title('A bounded non-linear function')
plt.axhline(lw=1, c='black')
plt.axvline(lw=1, c='black')
plt.grid(alpha=0.4, ls='-.')

plt.subplot(1, 3, 2)
plt.plot(x, y2, c='black', lw=3)
plt.title('A bounded non-linear function')
plt.axhline(lw=1, c='black')
plt.axvline(lw=1, c='black')
plt.grid(alpha=0.4, ls='-.')


plt.subplot(1, 3, 3)
plt.plot(x, y3, c='black', lw=3)
plt.axhline(lw=1, c='black')
plt.axvline(lw=1, c='black')
plt.grid(alpha=0.4, ls='-.')
plt.title('An unbounded non-linear function')


plt.show()
```
